```python
#df1
# 두 라벨의 가능한 모든 조합 만들기
combinations = np.array(np.meshgrid(df1.bias.unique(), df1.hate.unique())).T.reshape(-1,2)
# bias, hate 컬럼을 합친 것
bias_hate = list(np.array([df1['bias'].values, df1['hate'].values]).T.reshape(-1,2))
labels = []
for i, arr in enumerate(bias_hate):
    for idx, elem in enumerate(combinations):
        if np.array_equal(elem, arr):
            labels.append(idx)

df1['label'] = labels
df1 = df1.drop(['bias', 'hate'], axis=1)
df1.head()
```

```python
def more_than_half(nums):
  set_list = set(nums)
  result, count = 0, 0

  for i in set_list:
    if count < nums.count(i):
      result = i
      count  = nums.count(i)

  return result

answer = []
for i in range(len(df1)):
  nums = [df1['label'][i], df7['label'][i], df6['label'][i]]
  answer.append(more_than_half(nums))
```
